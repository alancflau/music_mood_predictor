## Data Exploration


Let's explore the music features

```python
music_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 
                  'acousticness', 'instrumentalness', 'liveness', 
                  'valence', 'tempo', 'duration_ms', 'time_signature']
```

```python
def min_max(x):
    """
    Determine the minimum and maximum values of a column
    
    """
    
    result = pd.Series(index =['min', 'max'], data = [x.min(), x.max()])
    
    return result
```

```python
my_songs[music_features].apply(min_max)
```

### Normalize features

Most music features are measured between 0 and 1, the exceptions are loudness and tempo. To reduce the misclassification when we fit the data into a model, we want to normalizing these two features.

```python
# Feature: loudness
loudness = my_songs[['loudness']]
min_max_scalar = preprocessing.MinMaxScaler()

loudness = min_max_scalar.fit_transform(loudness)
my_songs['loudness_norm'] = loudness

# Feature: tempo
tempo = my_songs[['tempo']]
tempo = min_max_scalar.fit_transform(tempo)
my_songs['tempo_norm'] = tempo

# # Feature: duration_ms
# duration_ms = my_songs[['duration_ms']]
# duration_ms = min_max_scalar.fit_transform(duration_ms)
# my_songs['duration_norm'] = duration_ms

# Feature: key
key = my_songs[['key']]
key = min_max_scalar.fit_transform(key)
my_songs['key_norm'] = key

# Feature: time_signature
time_signature = my_songs[['time_signature']]
time_signature = min_max_scalar.fit_transform(time_signature)
my_songs['time_signature_norm'] = time_signature
```

Add normalized features to the list of music of features

Based on rational thought, some of these features may not be useful in determining the mood of a song: features such as duration of the song.

I will make a new list of music featuers in which can be used to determine how many groups (clusters) we want in our mood prediction

```python
my_songs.columns
```

```python
new_music_features = ['danceability',
                      'energy',
                      'key_norm',
                      'mode',
                      'acousticness',
                      'instrumentalness',
                      'liveness',
                      'valence',
                      'tempo_norm',
                      'time_signature_norm',
                      'loudness_norm']

songs_features = my_songs[new_music_features]
```

### Cluster

Now that there is data, the data needs to be grouped so that the number of moods can be determined. Let's first find optimal k value

```python
pca = PCA(n_components = 0.95, random_state=42)
X_reduced = pca.fit_transform(songs_features)
X_reduced.shape
```

```python
# run kmeans with many different k
distortions = []
K = range(2, 15)
for k in K:
    k_means = KMeans(n_clusters = k, random_state=42).fit(X_reduced)
    k_means.fit(X_reduced)
    distortions.append(sum(np.min(cdist(X_reduced, k_means.cluster_centers_, 'euclidean'), axis=1)) / songs_features.shape[0])
    #print('Found distortion for {} clusters'.format(k))
```

```python
X_line = [K[0], K[-1]]
Y_line = [distortions[0], distortions[-1]]

# Plot the elbow
plt.figure(figsize=(15,6))
plt.plot(K, distortions, 'b-')
plt.plot(X_line, Y_line, 'r')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

From the plot above, the optimal k seems to be 5. After that, the decrease in distrotion is not as significant.

```python
k = 5
kmeans = KMeans(n_clusters = k, random_state=42)
y_pred = kmeans.fit_predict(X_reduced)
my_songs['labels'] = y_pred
```

```python
my_songs
```

```python
tsne = TSNE(verbose=1, perplexity=100, random_state=42)
X_embedded = tsne.fit_transform(X_reduced)
```

```python


# sns settings
sns.set(rc={'figure.figsize':(15,15)})

# colors
palette = sns.color_palette("bright", 1)

# plot
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], palette=palette)
plt.title('t-SNE with no Labels')
plt.savefig("t-sne_covid19.png")
plt.show()
```

```python
# sns settings
sns.set(rc={'figure.figsize':(15,15)})

# colors
palette = sns.hls_palette(5, l=.4, s=.9)

# plot
sns.scatterplot(X_embedded[:,0], X_embedded[:,1], hue=y_pred, legend='full', palette=palette)
plt.title('t-SNE with Kmeans Labels')
plt.savefig("improved_cluster_tsne.png")
plt.show()
```
